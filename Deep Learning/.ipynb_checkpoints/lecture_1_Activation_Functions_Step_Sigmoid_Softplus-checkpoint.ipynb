```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = np.linspace(-10, 10, 100)
```

# Step Function

```python
def step(x):
    return 1 if x >= 0 else 0
```

```python
y = [step(i) for i in x]
plt.plot(x, y)
plt.show()
```

# Sigmoid Function

```python
def sigmoid(x):
    return 1 / (1 + math.exp(-x))
```

```python
y = [sigmoid(i) for i in x]
plt.plot(x, y)
plt.show()
```

# Piecewise Function

```python
condlist = [x < 0, (x >= 0) & (x < 2), x >= 2]
funclist = [lambda x: -x, lambda x: x**2, 4]
```

```python
y = np.piecewise(x, condlist, funclist)
plt.plot(x, y)
plt.show()
```

# Tanh Function

```python
def tanh(x):
    return (2 / (1 + np.exp(-2*x)))
```

```python
y = [tanh(i) for i in x]
plt.plot(x, y)
plt.show()
```

# ReLU Function

```python
def relu(x):
    return x if x > 0 else 0
```

```python
y = [relu(i) for i in x]
plt.plot(x, y)
plt.show()
```

# Leaky Relu

```python
def leaky_relu(x, alpha=0.01):
    return max(x, alpha * x)
```

```python
y = [leaky_relu(i) for i in x]
plt.plot(x, y)
plt.show()
```

# Softmax

```python
def softmax(x):
    e_x = np.exp(x)
    return e_x / np.sum(e_x, axis=0)
```

```python
y = [softmax(i) for i in x]
plt.plot(x, y)
plt.show()
```

# Softplus Function

```python
def softplus(x):
    return np.log(1 + np.exp(x))
```

```python
y = [softplus(i) for i in x]
plt.plot(x, y)
plt.show()
```
