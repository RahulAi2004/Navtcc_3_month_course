```python
import math
import string
import sys
```

```python
def read_file(filename):
    try:
        with open(filename,'r',encoding='utf-8') as f:
            data=f.read()
        return data
    except IOError as e:
        raise IOError(f"Error opening or reading input file: {filename}") from e
```

```python
# Translation table: maps uppercase to lowercase, punctuation to spaces
translation_table = str.maketrans(
    string.punctuation + string.ascii_uppercase,
    " " * len(string.punctuation) + string.ascii_lowercase
)
```

```python
def get_words_from_line_list(text):
    text=text.translate(translation_table)
    word_list=text.split()
    return word_list
```

```python
def count_frequency(word_list):
    D={}
    for new_word in word_list:
        if new_word in D: D[new_word]+=1
        else: D[new_word]=1
    return D
```

```python
def word_frequencies_for_file(filename):
    line_list = read_file(filename)
    word_list = get_words_from_line_list(line_list)
    free_mapping = count_frequency(word_list)

    print(f"File {filename}:")
    print(f"    {len(line_list)} character, ")
    print(f"    {len(word_list)} words, ")
    print(f"    {len(free_mapping)} distinct words, ")

    return free_mapping
```

```python
def vector_angle(D1, D2):
  numerator = dotProduct(D1, D2)
  denominator = math.sqrt(dotProduct(D1, D1) * dotProduct(D2, D2))
  if denominator == 0:
    return math.pi / 2
  return math.acos(numerator / denominator)
```

```python
def dotProduct(D1, D2):
    Sum = 0.0
    for key in D1:
        if key in D2: Sum += D1[key] * D2[key]
    return Sum
```

```python
def documentSimilarity(file_1, file_2):
  sorted_word_list_1 = word_frequencies_for_file(file_1)
  sorted_word_list_2 = word_frequencies_for_file(file_2)
  distance = vector_angle(sorted_word_list_1, sorted_word_list_2)
  print(f"The distance between the documents is: {distance:0.6f} (radius)")
  return distance
```
