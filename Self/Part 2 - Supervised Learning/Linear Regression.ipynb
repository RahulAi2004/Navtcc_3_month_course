### How to actually create the best fit line?

y = mx + b

m = slope => x data point increase by 1 so how much y value will increase e.g. by 1

b = intercept => is the point you are starting y e.g m = 1, b = 0

x = data point to predict

y = prediction

e.g. if x = 1.5, m = 1, b = 0 find y

y = mx + b

y = 1(1.5) + 0 => 1.5

to create best fit line we continously change m and b value

e.g. x = 1.5, m = 0.5, b = 0

y = 0.5(1.5) + 0 => .75
