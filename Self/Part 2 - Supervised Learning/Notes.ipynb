R^2 = 1 - Sum of Residuals(Predictions after x_test) / Sum of Total
    => 1 - E (yi-y^i)^2 / E (yi-y)^2

### Overfitting Underfitting

if best fit line direcrly passes through the points its cost function is 0 perfect

if new data arrives it doesnt fit on line this problems shows model is overfit or underfit

when you train model it gives good result but in testing result isnt good

on training data model perform better but on testing the model not performed good

Overfitting

trining data -> performing better (low bias)

testing data -> model not performing good (high variance)


Underfitting

Means bad result in training and testing both

trining data -> performing bad (high bias)

testing data -> model not performing good (high variance)
