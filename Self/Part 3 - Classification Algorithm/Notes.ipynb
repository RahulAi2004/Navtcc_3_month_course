p(A and B) = P(A) * P(B |A) => given A also called conditional probability

Bayes theorem derives:

P(A and B) = P(B and A) => 

P(A) * P(B |A) = P(B) * P(A |B) => 

P(A|B) = P(A) * P(B|A) / P(B) => this is called bayes theorem that is derived by independent probability. 

independent(x1, x2, x3, x4) => y(dependent evt also called output)

P(y|x1 x2 x3 x4) = P(y) * P(x1 x2 x3 x4|y) / P(x1 x2 x3 x4)

### Probability of calculating yes =>

yes => P(y|x1 x2 x3 x4) = P(y) * P(x1|y) (x2|y) (x3|y) (xn|y) / P(x1 x2 x3 xn)

### Probability of calculating no =>

no => P(n|x1 x2 x3 x4) = P(n) * P(x1|n) (x2|n) (x3|n) (xn|n) / P(x1 x2 x3 xn)
